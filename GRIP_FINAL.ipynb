# Prediction using Supervised ML

## Predict the percentage of an student based on the no. of study hours.
● This is a simple linear regression task as it involves just 2 variables.
● Data can be found at http://bit.ly/w-data

## Importing required python libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Importing the dataset given via the link

```python
url = "http://bit.ly/w-data"
My_Data = pd.read_csv(url)
```

## Exploring the data provided

```python
My_Data.head()
```

```python
My_Data.tail()
```

```python
My_Data.shape
```

## Plotting certain graphs for better understanding and visualization

```python
My_Data.plot(x='Hours', y='Scores')  
plt.title('Hours - Percentage')  
plt.xlabel('Hours done of studying')  
plt.ylabel('Score in percentage')  
plt.show()
```

```python
My_Data.plot(x='Hours', y='Scores', style = 'o')  
plt.title('Hours - Percentage')  
plt.xlabel('Hours done of studying')  
plt.ylabel('Score in percentage')  
plt.show()
```

## Dividing data among x and y for training and testing data before applying ML Algos

```python
X = My_Data.iloc[:, :-1].values  
y = My_Data.iloc[:, 1].values  
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.25, random_state=0) 
```

## Using linear regression ML algorithm

```python
from sklearn.linear_model import LinearRegression  
```

```python
regression_model = LinearRegression()  
regression_model.fit(X_train, y_train) 
```

```python
print(X_test)
y_pred = regression_model.predict(X_test)
```

```python
mydf = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
mydf 
```

# What will be predicted score if a student studies for 9.25 hrs/ day?

```python
x_input = eval(input("Enter no of hr for prediction"))
predicated_value = regression_model.predict([[x_input]])
print(predicated_value)
```
